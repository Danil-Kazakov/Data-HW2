Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"



```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("2017_jun_final - 2017_jun_final.csv")

python_data = df[df['Мова програмування'] == 'Python']

aggregated_data = python_data.groupby('Посада').agg({'Зарплата на місяць': ['min', 'max']})

aggregated_data.columns = ['Мінімальна зарплата', 'Максимальна зарплата']

print(aggregated_data)

aggregated_data.reset_index(inplace=True)

# Побудова графіка
plt.figure(figsize=(12, 8))

# Графік мінімальних зарплат
plt.barh(aggregated_data['Посада'], aggregated_data['Мінімальна зарплата'], color='lightblue', label='Мінімальна зарплата')

# Графік максимальних зарплат
plt.barh(aggregated_data['Посада'], aggregated_data['Максимальна зарплата'], color='skyblue', label='Максимальна зарплата')

plt.xlabel('Зарплата на місяць, USD')
plt.ylabel('Посада')
plt.title('Діапазон мінімальних та максимальних зарплат програмістів на Python за посадами')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
