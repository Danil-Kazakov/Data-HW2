Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
import pandas as pd
source = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C'
df = pd.read_html(source, match='Коефіцієнт народжуваності в регіонах України', thousands=".", decimal=",")[0]
df.replace('—', np.nan, inplace=True)
df['2019'] = pd.to_numeric(df['2019'], errors='coerce')
df = df.drop(df.index[-1])
average_birth_rate_2019 = df['2019'].mean()
regions_above_average = df[df['2019'] > average_birth_rate_2019]['Регіон'].tolist()
print(regions_above_average)
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

source = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C'
df = pd.read_html(source, match='Коефіцієнт народжуваності в регіонах України', thousands=".", decimal=",")[0]
df.replace('—', np.nan, inplace=True)
df['2014'] = pd.to_numeric(df['2014'], errors='coerce')
region_max_birth_rate_2014 = df.loc[df['2014'].idxmax(), 'Регіон']
print(region_max_birth_rate_2014)

plt.figure(figsize=(12, 8))
plt.barh(df['Регіон'], df['2014'], color='skyblue')
plt.xlabel('Коефіцієнт народжуваності')
plt.ylabel('Регіон')
plt.title('Коефіцієнт народжуваності в регіонах України за 2014 рік')
plt.grid(True, axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
source = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C'
df = pd.read_html(source, match='Коефіцієнт народжуваності в регіонах України', thousands=".", decimal=",")[0]
df.replace('—', np.nan, inplace=True)
df['2019'] = pd.to_numeric(df['2019'], errors='coerce')

plt.figure(figsize=(12, 8))
plt.bar(df['Регіон'], df['2019'], color='skyblue')
plt.xlabel('Регіон')
plt.ylabel('Народжуваність у 2019 році')
plt.title('Народжуваність по регіонах у 2019 році')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
